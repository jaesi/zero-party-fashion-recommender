```python
print("행 x 열 사이즈 :", df.shape)
print("데이터 형식:", df.dtypes)
# 450 x 110 데이터프레임
```

```python
#fillna() NaN 결측치 값을 0으로 바꿔주는 것
df = df.fillna(0)
```

```python
# 1 MBTI Binary
# ESTP -> 1로

mbti = ['E/I', 'S/N', 'T/F', 'P/J']


for col in mbti:
    df[col] = df[col].apply(lambda x: 1 if x == 'E' else 
                            (1 if x == 'S' else 
                             (1 if x == 'T' else 
                              (1 if x == 'P' else 
                               0))))


df.head()
```

```python
# 결과 열 따로 분리 
# q11_1~8, q12_1~9

result_columns = ['q11_1', 'q11_2', 'q11_3', 'q11_4', 'q11_5', 'q11_6', 'q11_7', 'q11_8'] 
                  # 'q12_1', 'q12_2', 'q12_3', 'q12_4', 'q12_5', 'q12_6', 'q12_7', 'q12_8', 'q12_9']



X = df.drop(columns=result_columns)
y = df[result_columns]
```

```python
y
```

```python
# 원핫 인코딩 재결합

y['socks'] =y.idxmax(axis=1)
y.drop(y.columns[:-1], axis=1, inplace=True)
y
```

```python
y.socks.unique()
```

```python
# 결과값 라벨 인코딩 

for i in range(1,9):
    y['socks'] = y['socks'].replace(f'q11_{i}', int(f'{i}'))

print(y)
print(y.dtypes)
```

```python
plt.figure(figsize=(10,8))
plt.hist(y)
# plt.savefig('train_results.png')
plt.show()
```

```python
# 학습용과 테스트용으로 데이터 분리
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=50)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

# 결정 트리 분류기
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
# 트레이닝
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
y_pred = dt.predict(X_test)
rf_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy = dt.score(X_test, y_test)
accuracy_rf = rf.score(X_test, y_test)
print('DT Accuracy :', accuracy)
print('rf Accuracy :', accuracy_rf)
```
